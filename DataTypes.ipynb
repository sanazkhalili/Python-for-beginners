#Integer

---
for example : 
1, -3, 2356, 99999

|   |   |   |
|---|---|---|
| +  |  2 + 3 |  
| -  | 2 - 3  |  
| *  |  2 * 3 |  
|  ** | 2 ** 3  |
| //  |  2 // 3 |
|  % | 2 % 3  |
|  / | 2 / 3  |


---




```python
print('2 + 3 = ', 2 + 3)
print('2 - 3 = ', 2 - 3)
print('2 * 3 = ', 2 * 3)
print('2 // 3 = ', 2 // 3)
print('2 % 3 = ', 2 % 3)
print('2 / 3 = ', 2 / 3)
```

#Float

#Boolean
True or False




| Operator | True  | False |
|---|---|---|
| ==  |  5==5 | 5==3  |
| !=  | 3!=2  | 2!=2  |
| >  |  2>1 | 2>3  |  
|  < | 2<3  | 4<3  |  
| <=  |  2<=2 | 2<=1  |   
| >=  |  6>=5 | 6>=7  |   


```python
flag1 = False
flag2 = True
print("flag1 and flag2: " + str(flag1 and flag2))
print("flag1 or flag2: " + str(flag1 or flag2))
```
